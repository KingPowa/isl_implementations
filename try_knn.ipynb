```python
# KNN Test
import numpy as np
import matplotlib.pyplot as plt

from isl.knn import KNNClassification
```

```python
def exponential_boundary(x: np.ndarray) -> np.ndarray:
    return np.power(x, 2)
```

```python
x_train = np.random.rand(300, 2)
y_train = (x_train[:, 1] > exponential_boundary(x_train[:, 0])).astype(np.int8)

fig = plt.figure(figsize = (6,6))
fig.suptitle('True boundary with train points')
plt.scatter(x_train[y_train == 0][:, 0], x_train[y_train == 0][:, 1], color='blue', label='y_train = 0', alpha=0.6)
plt.scatter(x_train[y_train == 1][:, 0], x_train[y_train == 1][:, 1], color='red', label='y_train = 1', alpha=0.6)

# Boundary
boundary_x = np.linspace(0, 1, 300)
boundary_y =  exponential_boundary(boundary_x)
plt.plot(boundary_x, boundary_y)
```

```python
classifier = KNNClassification(30)
classifier.fit(x_train[:250], y_train[:250])
```

```python
y_test = classifier.predict(x_train[250:])
y_test
```

```python
fig = plt.figure(figsize = (6,6))
fig.suptitle('True boundary with test points')
plt.scatter(x_train[250:, 0][y_test == 0], x_train[250:, 1][y_test == 0], color='blue', label='y_train = 0', alpha=0.6)
plt.scatter(x_train[250:, 0][y_test == 1], x_train[250:, 1][y_test == 1], color='red', label='y_train = 1', alpha=0.6)

# Boundary
boundary_x = np.linspace(0, 1, 300)
boundary_y = exponential_boundary(boundary_x)
plt.plot(boundary_x, boundary_y)
```
